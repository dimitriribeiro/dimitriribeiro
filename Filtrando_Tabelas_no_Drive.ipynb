```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

# Filtrando Semana 1

```python
#Importando o HTML para iniciar a raspagem a partir do link da tabela

url = requests.get('https://docs.google.com/spreadsheets/d/e/2PACX-1vTxsFzH6-kujfjj8a6SLWfd5XktxHzuKyHIGtW8V_IsdDmqZMf5KPHlhTKMfL3Z97HTK98y5kjyOgEj/pubhtml')
html = url.text
soup = BeautifulSoup(html, 'html.parser')

#Iniciando filtragem, pegando apenas as células da tabela que contenham texto pela TAG <td>

semana1 = []
for linha in soup.find_all('td'):
  if linha.text == '':
    pass
  else:
    semana1.append(linha.text)

#Separando conteúdo adqurido em duas categorias, código e denominação

codigos1 = []
denominacao1 = []
for elemento in semana1:
  if elemento.isnumeric():
    codigos1.append(elemento)
  else:
    denominacao1.append(elemento)

#OBS: Nessa tabela, tivemos que tirara celulas que ocupavam mais de um espaço em cada subtabela

dispensaveis = ['UTENSÍLIOS PLÁSTICOS E SIMILAR', 'UTENSÍLIOS DE AÇO E SIMILAR', 'PROTEÇÃO E SEGURANÇA','MATERIAL DE COZINHAR E PASSAR', 'LOUÇAS E ARTIGOS DE MESA']
#Então, retiramos elas com um for

for inviavel in denominacao1:
  if inviavel in dispensaveis:
    denominacao1.remove(inviavel)
#Agora, retirar elementos das nossas categorias que não representam valor para o que queremos

denominacao1 = denominacao1[40:]
codigos1 = codigos1[17:]

#Formando DataFrame a partir de um dicionário. Recomendo.
semana1_data = {'CÓDIGOS':codigos1,'DENOMINAÇÃO':denominacao1}

dataset_semana1 = pd.DataFrame(semana1_data)

#Obs: Foi solicitado que adicionasse mais uma coluna no nossa DataFrame

dataset_semana1['GRUPOS'] = pd.Series()

#Utilizamos os nomes das seções da tabela original para categorizar por código
#Obs: As listas nome e código estão em ordem e tem mesmo tamanho, se a ordem for alterada, a tabela não será produzida.


nomes = ['MATERIAL QUIMICO', 'MATERIAL BIOLÓGICO','GENEROS DE ALIMENTACAO','MATERIAL DE CAMA, MESA E BANHO','MATERIAL DE COPA E COZINHA', 'MATERIAL DE ACONDICIONAMENTO E EMBALAGEM', 'MATERIAL DE COUDELARIA OU DE USO ZOOTÉCNICO', 'MATERIAL DE PROCESSAMENTO DE DADOS']
codigo = ['3011','3040','3007','3020','3021','3019','3012','3017']
for i, cod in enumerate(dataset_semana1['CÓDIGOS']):
  if codigo[0] in cod:
    dataset_semana1['GRUPOS'][i] = nomes[0]
  elif codigo[1] in cod:
    dataset_semana1['GRUPOS'][i] = nomes[1]
  elif codigo[2] in cod:
    dataset_semana1['GRUPOS'][i] = nomes[2]
  elif codigo[3] in cod:
    dataset_semana1['GRUPOS'][i] = nomes[3]
  elif codigo[4] in cod:
    dataset_semana1['GRUPOS'][i] = nomes[4]
  elif codigo[5] in cod:
    dataset_semana1['GRUPOS'][i] = nomes[5]
  elif codigo[6] in cod:
    dataset_semana1['GRUPOS'][i] = nomes[6]
  else:
    dataset_semana1['GRUPOS'][i] = nomes[7]

#Filtração para aquisição de tabela editável

indices_semana1 = []
for i, codigo in enumerate(dataset_semana1['CÓDIGOS']):
  if '3017' in codigo or '3019' in codigo or '3040' in codigo or '3011' in codigo:
    indices_semana1.append(i)
dataset_semana1 = dataset_semana1.iloc[indices_semana1]
dataset_semana1 = dataset_semana1.sort_values(by='GRUPOS')
#Exportando para Excel Tabela Filtrada e Categorizada
dataset_semana1.to_excel('Semana 1_Produtos_ATUALIZADO.xlsx')
```
